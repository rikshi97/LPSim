```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = pd.read_csv('random_sets.csv').iloc[:125,:]
split = x['Numbers'].str.split(",")
x = np.zeros((125, 20))
for i in range(125):
    idx = np.array(split[i], dtype=int)-1
    x[i,idx] = 1
y = pd.read_csv('total_travel_time.csv')['total_travel_time'].to_numpy()
```

```python
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize=(4,8))
sns.heatmap(x, 
            cmap='Blues', 
            yticklabels=False,
            cbar=False,
            ax=ax)
ax.set(title='Sampled Vertiport Locations', 
       xlabel='Vertiport Index', 
       ylabel='Sample Index')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
sns.histplot(y/1000000)
plt.title('Cost Distributions')
plt.xlabel('Total Travel Time (M)')
```
